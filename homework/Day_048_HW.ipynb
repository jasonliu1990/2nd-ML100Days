```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('dataset/train.csv',header=None)
y = pd.read_csv('dataset/trainLabels.csv',header=None)
test = pd.read_csv('dataset/test.csv',header=None)
```

```python
print(train.shape)
print(y.shape)
print(test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
train = std.fit_transform(train)
test = std.fit_transform(test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=15)
train_pca = pca.fit_transform(train)
test_pca = pca.fit_transform(test)
```

```python
train.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_Y, val_Y = train_test_split(train_pca, y, test_size=0.2, random_state=234)
```

```python
print(train_X.shape, train_Y.shape)
print(val_X.shape, val_Y.shape)
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score
```

```python
train_data = lgb.Dataset(train_X, label=train_Y)
validation = lgb.Dataset(val_X, label=val_Y)
```

```python
params={
    'boosting_type': 'gbdt',
    'objective': 'binary',
    
    'learning_rate': 0.01,
    'max_depth': 8,
    'num_leaves': 80,
    'subsample': 0.8,
    'min_data_in_leaf':18,
    
    'n_estimators': 1000,
#     'reg_alpha':0.2,
    'reg_lambda':0.6,

    
}
```

```python
model = lgb.train(params, train_data, valid_sets=[validation])
```

```python
val_pred = model.predict(val_X)
print(accuracy_score(val_Y, list(map(lambda x: round(x,0), val_pred))))
```

```python
pred = model.predict(test_pca)
pred = list(map(lambda x: int(round(x,0)), pred))
```

```python
sub = pd.DataFrame()
sub['Id'] = list(range(1,9001))
sub['Solution'] = pred
```

```python
sub.to_csv('sub_1.csv', index=False)
```
